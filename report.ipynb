```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('../eval_summary.csv')  # Replace with your actual file path
print(df.columns)
# Set the plot style
sns.set(style="whitegrid")

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='method', y='percentage', data=df, palette='coolwarm')

# Set plot labels and title
plt.xlabel('Method')
plt.ylabel('Match Percentage')
plt.title('Match Percentage by Method')
plt.xticks(rotation=45, ha='right')
plt.ylim(0, 1)  # Assuming the percentage is between 0 and 1
plt.tight_layout()

# Show the plot
plt.show()
```
